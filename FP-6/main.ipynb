### Purpose:

* The purpose of this project is to analyze the factors that contribute to airline flight delays and cancellations in the United States. This analysis will be conducted with data from US domestic flights in the year 2024.

* The key **dependent variables** (DVs) is:
    - ArrDelay (Arrival Delay in minutes)

* Key **independent variables** (IVs) include:
    - DepDelay (Departure delay in minutes)
    - DayOfWeek
    - CRSDepTime (Computer Reservation System Departure Time)
    - Reporting_Airlines (Airline Carrier Code)
    - Dest
    - Distance
    - Month

* This dataset contains data from all major US domestic flights in 2024. The dataset includes 1 primary dependent variable (arrival in minutes) and 7 key independent variables. The full dataset contains 110 total variables. In the GitHub under dataReadme.md further description of all of this variables provided by BTS can be found. In total there are 7,079,061 entries in the dataset, this does not mean all flights in 2024 in the US, just the ones tracked by BTS. For the sake of this project only the flights departing from Philadelphia International Airport (PHL) will be analysed. This means the dataset goes from 7M to 100k flights/entries.

<br>

### Dataset source:

The data comes from the **Bureau of Transportation Statistics (BTS)** - [Airline On-Time Performance Data](https://www.transtats.bts.gov/DL_SelectFields.aspx?gnoyr_VQ=FGJ&QO_fu146_anzr=b0-gvzr).

This is maintained by the U.S. Department of Transportation and provides data on flights operated by large air carriers.

We will use data from 2024 (January-December)
