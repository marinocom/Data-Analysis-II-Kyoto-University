### Purpose:

* The purpose of this project is to analyze the factors that contribute to airline flight delays and cancellations in the United States. This analysis will be conducted with data from US domestic flights in the year 2024.

* The key **dependent variables** (DVs) are:
    * ArrDelay - Arrival delay in minutes (can be negative for early arrivals)
    * DepDelay - Departure delay in minutes
    * Cancelled - Binary: 1 if flight cancelled, 0 otherwise
    * Diverted - Binary: 1 if flight diverted, 0 otherwise

* Key **independent variables** (IVs) include:
    * Temporal factors:
        * Month - Month of the year (1-12)
        * DayofMonth - Day of the month (1-31)
        * DayOfWeek - Day of the week (1-7, where 1=Monday)
        * CRSDepTime - Scheduled departure time
        * CRSArrTime - Scheduled arrival time
    * Airline characteristics:
        * Reporting_Airline - Airline carrier code
        * Tail_Number - Aircraft tail number
        * Flight_Number_Reporting_Airline - Flight number
    * Airport and route characteristics:
        * Origin - Origin airport code
        * Dest - Destination airport code
        * OriginState - State of origin airport
        * DestState - State of destination airport
        * Distance - Distance between origin and destination airports (miles)
        * DistanceGroup - Categorical distance grouping
    * Flight scheduling:
        * CRSElapsedTime - Scheduled flight duration (minutes)
        * DepTimeBlk - Departure time block (categorical)
        * ArrTimeBlk - Arrival time block (categorical)

* This dataset contains data from all major US domestic flights in 2024. The dataset includes 4 primary dependent variables (arrival delay, departure delay, cancellation status, and diversion status). There are approximately 20-25 key independent variables across temporal, airline, airport, and flight scheduling categories. The full dataset contains 110 total variables. In the GitHub under dataReadme.html further description of all of this variables provided by BTS can be found. In total there are 7,079,061 entries in the dataset, this does not mean all flights in 2024 in the US, just the ones tracked by BTS.

### Dataset source:

The data comes from the **Bureau of Transportation Statistics (BTS)** - [Airline On-Time Performance Data](https://www.transtats.bts.gov/DL_SelectFields.aspx?gnoyr_VQ=FGJ&QO_fu146_anzr=b0-gvzr).

This is maintained by the U.S. Department of Transportation and provides data on flights operated by large air carriers.

We will use data from 2024 (January-December)
