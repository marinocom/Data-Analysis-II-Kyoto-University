### Figure 2

Use `data-fig2.npy` for this figure.

```python
# Load data for Figure 2
b = np.load('data-fig2.npy')

# Create line plot with three series in red, green, and blue
x = np.arange(len(b))

plt.figure(figsize=(8, 6))
plt.plot(x, b[:, 0], color='red', linewidth=2, label='Red')
plt.plot(x, b[:, 1], color='green', linewidth=2, label='Green')
plt.plot(x, b[:, 2], color='blue', linewidth=2, label='Blue')
plt.xlabel('Y value', fontsize=12)
plt.ylabel('X value', fontsize=12)
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

___

### Figure 3

Use `data-fig3.npy` for this figure.

```python
# Load data for Figure 3
c = np.load('data-fig3.npy')

# Create scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(c[:, 0], c[:, 1], s=175, alpha=0.3, c='blue', edgecolors='darkblue', linewidth=1)
plt.xlabel('X value', fontsize=12)
plt.ylabel('Y value', fontsize=12)
plt.show()
```

___

### Figure 4

Use `data-fig4.npy` for this figure.

```python
# Load data for Figure 4
d = np.load('data-fig4.npy')

# Create bar chart
x = np.arange(len(d))

plt.figure(figsize=(12, 10))
plt.bar(x, d, color='gray', edgecolor='gray', linewidth=1)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.xticks(x, [chr(65 + i) for i in range(len(d))])
plt.tight_layout()
plt.show()
```

### Figure 5

Use `data-fig5.npy` for this figure.

*Hint*: Use 2D visualization and the "RdBu" colormap.

```python
# Load data for Figure 5
e = np.load('data-fig5.npy')

# Create 2D heatmap with RdBu colormap
plt.figure(figsize=(12, 10))
im = plt.imshow(e, cmap='RdBu', aspect='auto', origin='lower', vmin=-1.0, vmax=1.0)
plt.colorbar(im)
plt.tight_layout()
plt.show()
```

### UNGRADED BONUS QUESTION

(This question is an *optional*, *ungraded* bonus problem.) 

Use `data-fig-bonus-X.npy`, `data-fig-bonus-Y.npy` and `data-fig-bonus-Z.npy` for this figure.

*Hint*: Search Matplotlib's online examples.

```python
# Load data for UPGRADED BONUS QUESTION figure
X = np.load('data-fig-bonus-X.npy')
Y = np.load('data-fig-bonus-Y.npy')
Z = np.load('data-fig-bonus-Z.npy')

fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# Plot 3D contour with extend3d to create the layered spiral effect
ax.contour3D(X, Y, Z, extend3d=True, cmap='coolwarm')

ax.grid(True) # Add grid
plt.tight_layout()
plt.show()
```
